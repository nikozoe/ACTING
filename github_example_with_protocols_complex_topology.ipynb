## Model github group with star topology

```python
from ACTING_base import *
import itertools
import networkx as nx
from nxpd import draw
import numpy as np
from mesa.batchrunner import BatchRunner
import pandas as pd
import networkx
import matplotlib
import matplotlib.pyplot as plt
```

#### Load EPA values of identities we use in the simulations

```python
#identities

github_convinience_survey = pd.read_csv("GitHub_Convenience_Sample_17April2019.csv")
developer = github_convinience_survey[["developer_e_1","developer_p_1","developer_a_1"]].mean().values

submit_pr = github_convinience_survey[["pull-request_b_e_1","pull-request_b_p_1","pull-request_b_a_1"]].mean().values
accept_pr = github_convinience_survey[["acc_pull-request_e_1","acc_pull-request_p_1","acc_pull-request_a_1"]].mean().values
reject_pr = github_convinience_survey[["rej_pull-request_e_1","rej_pull-request_p_1","rej_pull-request_a_1"]].mean().values
novice_developer = github_convinience_survey[["novice_developer_e_1","novice_developer_p_1","novice_developer_a_1"]].mean().values
experienced_developer = github_convinience_survey[["exp_developer_e_1","exp_developer_p_1","exp_developer_a_1"]].mean().values
```

#### Some helper functions for data processing and visualization

```python
#helper class, because variable parameters passed to Batchrunner have to be hashable
class HashableList(list):
    def __hash__(self):
        return hash(str(self))

    
def agent_data_collector(model):
    return model.datacollector.get_agent_vars_dataframe()

def model_data_collector(model):
    return model.datacollector.get_model_vars_dataframe()

def totuple(a):
    try:
        return tuple(totuple(i) for i in a)
    except TypeError:
        return a

def group_by_id(df):
    grouped=df.groupby("AgentID")
    new = pd.concat([grouped.get_group(i).rename(columns=lambda x: x + str(i)).reset_index(drop=True) for i in range(len(grouped))], axis=1)
    return new

def epa_means(df, agents):
    agentdf = df.loc[df["agents"].apply(lambda x: str(x) == str(agents))]["agent data"]
    agentdf=agentdf.apply(lambda x: group_by_id(x))
    zwi = pd.concat([f for f in agentdf])
    by_row_index = zwi.groupby(zwi.index)
    df_means = by_row_index.mean()
    return df_means

def epa_means_shuffled(df):
    agentdf = df["agent data"]
    agentdf=agentdf.apply(lambda x: group_by_id(x))
    zwi = pd.concat([f for f in agentdf])
    by_row_index = zwi.groupby(zwi.index)
    df_means = by_row_index.mean()
    return df_means


def model_means(df, agents):
    agentdf = df.loc[df["agents"].apply(lambda x: str(x) == str(agents))]["agent data"]
    zwi = pd.concat([f for f in agentdf])
    by_row_index = zwi.groupby("Step")
    df_means = by_row_index.mean()
    return df_means

def model_means_shuffled(df):
    agentdf = df["agent data"]
    zwi = pd.concat([f for f in agentdf])
    by_row_index = zwi.groupby("Step")
    df_means = by_row_index.mean()
    return df_means


def draw_github_project(net,node_color, int_labels=False, edge_label=False):
    testnw=net
    if int_labels:
        testnw= nx.convert_node_labels_to_integers(testnw)
    nx.set_node_attributes(testnw, node_color ,'color')
    nx.set_node_attributes(testnw, 2.0 ,'penwidth')
    testnw.graph['dpi'] = 80
    if edge_label == True:
        for pairs in testnw.adjacency():
            for edge in pairs[1]:   
                testnw[pairs[0]][edge]['label']=str(pairs[1][edge]['weight'])
    fig=plt.figure(figsize=(10,2))
    
    plt.tight_layout()
    for (n1, n2, d) in testnw.edges(data=True):
        d.clear()
    print(testnw.edges)
    return draw(testnw, show='ipynb')
            
colors=["#35a8e0", "#05c100", "#ffd300",'#d8b69f', '#d1000a']
epa_colors=["#cb9f00", "#058c7d", "#78325f"]
```

#### draw network topology

```python
G=nx.read_adjlist("complexcluster_example.adjlist", create_using=nx.DiGraph)
draw_github_project(G, colors[1], int_labels=True, edge_label=False)
```

In the example graph the directed edges point from the developer who submits to the
developer who accepts the PR. But when submitting a PR, a developer does not necessarily know in advance
who is going to review it. Hence, to simulate the network structure, we reverse all edges because accepting a PR
is without a doubt a directed action towards the submitter while submitting a PR is more accurately modeled as
an action on the whole group. We also remove self loops, since an action on oneself does not have a sound interpretation in Affect Control Theory.

```python
#remove self loops
loops = G.selfloop_edges()
G.remove_edges_from(loops)

#reverse links and transform to tuple to make it hashable so that the mesa Batchrunner can handle it as input
adj_tuple = totuple(nx.to_numpy_array(G.reverse(), weight=None))
```

### prepare agent sets:
We run 4 different experiments with 4 different agent sets:

a) (agentset1) All agents are initialized with the fundamental identity of "experienced developer".

b) (agentset2) All agents are initialized with the fundamental identity of "novice developer".

c) (agentset3) All agents who have higher status and can accept PRs ( i.e. who have an incoming edge in the Figure above) are initialized with the fundamental identity of "experienced developer". All other agents are initialized with the fundamental identity "novice developer".

d) (agent_variations_shuffled) Agents are randomly initialized with either the identity "experienced developer" or "novice" independent of their network position, but with the same numbers of experienced/novice developers as
in setup c). 

```python
np.random.seed(seed=123)

agents_variations_github = []

agentset1 = [{"epa": novice_developer, "initial_tension": 1.0, "individuality": 1.0} for node in adj_tuple]

agentset2 = [{"epa": experienced_developer, "initial_tension": 1.0, "individuality": 1.0} for node in adj_tuple]


agentset3 = []
degrees = np.sum(adj_tuple, axis=1)
for degree in degrees:
    if degree > 0:
        agentset3.append({"epa": experienced_developer, "initial_tension": 1.0, "individuality": 1.0})
    else:
        agentset3.append({"epa": novice_developer, "initial_tension": 1.0, "individuality": 1.0})
        
    
agents_variations_github = [HashableList(agentset1), HashableList(agentset2), HashableList(agentset3)]
```

### Definition of the task group protocol
1. An agent
submits a PR which is interpreted as an action on the group.

2. The group discusses the PR. This is simulated
via a sequence of freely chosen actions that are performed along the network topology or on the group object.
The length of the sequence is determined randomly to be between 0 and 10 actions. 

3. An agent with merge
rights who shares a link with the submitter (specified via the network topology), either accepts or rejects the
PR. The agent that actually gets to act is the one that experiences the most deflection among the candidates
that come into consideration.

To mimick this behavior we have to write some additional custom code that is parsed to the ProtocolGroupModel.
(See the doc string in ProtocolGroupModel for more info on how this is implemented in group_simulator_base_75_maxdeffix.py)

```python

submitter_select = """
self.reciprocal = False
possible_actors = []
for ag in self.schedule.agents:
    if self.network_structure[:,ag.unique_id].sum()>=1:
        possible_actors.append(ag)
np.random.shuffle(possible_actors)
self.actor = max(possible_actors, key=lambda ag: ag.personal_deflection)
self.actor.acting = True
"""


pr_select = """
self.reciprocal = True
self.object = self.actor
possible_actors = []
for ag in self.schedule.agents: 
    if self.network_structure[ag.unique_id, self.object.unique_id] == 1:
        possible_actors.append(ag)
np.random.shuffle(possible_actors)
self.actor = max(possible_actors, key=lambda ag: ag.personal_deflection)
self.actor.acting = True
"""

comment_code_bef = """
self.temp_object = self.object
self.temp_actor = self.actor
"""
comment_code_aft = """
self.object = self.temp_object
self.actor = self.temp_actor
"""

protocol = [
    {"discrete_actions":[submit_pr], "select_actor": submitter_select, "action_on_group": True, "group_action_rate": 1.0, "n_iterations": 1},
    {"discrete_actions":None, "code": (comment_code_bef, comment_code_aft),"group_action_rate": 0.0, "n_iterations": [0,10]},
    {"discrete_actions": [accept_pr, reject_pr], "select_actor": pr_select,"group_action_rate": 0.0, "n_iterations": 1}
]
```

#### Run simulations for experiments a), b) and c) (agentset1, agentset2 and agentset3)

```python
fixed_params = {
                "data_model": "us_unisex",
                "actor_choice": "max self-tension",
                "object_choice": "max deflection-reduction",
                "action_on_group": True,
                "group_action_rate": 0.0,
                "reciprocity_rate" : 0.0,
                "network_structure" : adj_tuple,
                "protocol": HashableList(protocol)
    
               }

variable_params = { 
                   "agents": agents_variations_github,
                  }




batch_run = BatchRunner(ProtocolGroupModel,
                        fixed_parameters=fixed_params,
                        variable_parameters=variable_params,
                        iterations=5000,
                        max_steps=100,
                        model_reporters={"model data": model_data_collector, "agent data": agent_data_collector}  
                       )
                      
batch_run.run_all()
model_data = batch_run.get_model_vars_dataframe()
```

#### Run simulations for experiments d) (agent_variations_shuffled)

```python
n = 5000
agent_variations_shuffled = []
for i in range(n):
    shuffled = agentset3.copy()
    np.random.shuffle(shuffled)
    agent_variations_shuffled.append(HashableList(shuffled))
```

```python
fixed_params = {
                "data_model": "us_unisex",
                "actor_choice": "max self-tension",
                "object_choice": "max deflection-reduction",
                "action_on_group": True,
                "group_action_rate": 0.0,
                "reciprocity_rate" : 0.0,
                "network_structure" : adj_tuple,
                "protocol": HashableList(protocol)
               }
variable_params = { 
                   "agents": agent_variations_shuffled
                  }
                  

batch_run_shuffled = BatchRunner(ProtocolGroupModel,
                        fixed_parameters=fixed_params,
                        variable_parameters=variable_params,
                        iterations=1,
                        max_steps=100,
                        model_reporters={"model data": model_data_collector, "agent data": agent_data_collector}  
                       )
                      
batch_run_shuffled.run_all()

model_data_shuffled = batch_run_shuffled.get_model_vars_dataframe()
```

```python
markers_array = list(matplotlib.markers.MarkerStyle.markers.keys())
lines_array = list(matplotlib.lines.lineStyles.keys())
line_styles = lines_array[:4] + markers_array
params = {'legend.fontsize': 14,
          'figure.figsize': (15, 5),
         'axes.labelsize': 16,
         'axes.titlesize':18,
         'xtick.labelsize':14,
         'ytick.labelsize':14,
         'font.family': "serif",
         'font.serif': 'Source Sans Pro'}
matplotlib.rcParams.update(params)


fig = plt.figure(figsize=(20, 4.8))
ax1 = fig.add_subplot(141)
ax2 = fig.add_subplot(142)
ax3 = fig.add_subplot(143)
ax4 = fig.add_subplot(144)

ax1.set_ylim(7.5, 12.5)
ax2.set_ylim(1.0, 6)
ax3.set_ylim(1.0, 6)
ax4.set_ylim(1.0, 6)

# Experienced developers
df_means1 = epa_means(model_data, agents_variations_github[1])
df_means1[["Deflection" + str(j) for j in range(len(adj_tuple))]][0:50].plot(
    title="a) Experienced developers",
    style=line_styles[: len(adj_tuple)],
    color="#35a8e0",
    ax=ax1,
    legend=False,
)
abs_means1 = model_means(model_data, agents_variations_github[1])
abs_means1["Deflection"][0:50].plot(
    style="--", color=colors[4], ax=ax1, label="Mean Deflection", legend=False
)

# Novice developers
df_means2 = epa_means(model_data, agents_variations_github[0])
df_means2[["Deflection" + str(j) for j in range(len(adj_tuple))]][0:50].plot(
    style=line_styles[: len(adj_tuple)],
    title="b) Novice developers",
    color="#35a8e0",
    ax=ax2,
    legend=False,
)
abs_means2 = model_means(model_data, agents_variations_github[0])
abs_means2["Deflection"][0:50].plot(
    style="--", color=colors[4], ax=ax2, label="Mean Deflection", legend=False
)

# Exp/Novice assigned according to network position
df_means3 = epa_means(model_data, agents_variations_github[2])
df_means3[["Deflection" + str(j) for j in range(len(adj_tuple))]][0:50].plot(
    style=line_styles[: len(adj_tuple)],
    title="c) Exp./Nov. matching position",
    color="#35a8e0",
    ax=ax3,
    legend=False,
)
abs_means3 = model_means(model_data, agents_variations_github[2])
abs_means3["Deflection"][0:50].plot(
    style="--", color=colors[4], ax=ax3, label="Mean Deflection", legend=False
)

# Exp/Novice assigned randomly to network positions
df_means4 = epa_means_shuffled(model_data_shuffled)
df_means4[["Deflection" + str(j) for j in range(len(adj_tuple))]][0:50].plot(
    title="d) Exp./Nov. assigned randomly",
    style=line_styles[: len(adj_tuple)],
    color="#35a8e0",
    ax=ax4,
    legend=False,
)
abs_means4 = model_means_shuffled(model_data_shuffled)
abs_means4["Deflection"][0:50].plot(
    style="--", color=colors[4], ax=ax4, label="Mean Deflection", legend=False
)
ax4.legend(ncol=2, bbox_to_anchor=(1.0, 1.0))

ax1.set_xlabel("steps")
ax2.set_xlabel("steps")
ax3.set_xlabel("steps")
ax4.set_xlabel("steps")
plt.show()
```
